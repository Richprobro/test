**Data Preparation**

```python
#i will perform a filter to get people with age more than 30
filtered_database = db[db['Age'] > 30]
print("Filtered db (Age > 30):")
```

```python
# counting gender that starts with letter M which stands for male
letter = 'M'
matching_letters = db[db['Gender'].str.startswith(letter)]
print(f"Total records where 'Gender' starts with '{letter}':", matching_letters.shape[0])
```

```python
# identifying duplicates and removing them
duplicate_rows = db.duplicated().sum()
print("Total duplicate rows:", duplicate_rows)
```

```python
db_cleaned = db.drop_duplicates()
print("Dataset after removing duplicates:", db_cleaned.shape)
```

```python
column = 'Age'
db[numerical_column] = db[column].astype(str)
print(f"Converted '{column}' to string. New data type:")
print(db.dtypes[column])
```

```python
# grouping the data based on gender and history of obesity of families
grouped_data = db.groupby(['Gender', 'family_history_with_overweight']).size().reset_index(name='Count')
print("Grouped dataset analysis:")
print(grouped_data)
```

```python
# Checking for missing values(null ones)
missing_values = db.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# replacing misiing values with mode or median values but no need as there is no null values in our chosen dataset
for column in db.columns:
    if db[column].isnull().sum() > 0:
        if db[column].dtype == 'object':  # the categorical values
            db[column].fillna(db[column].mode()[0], inplace=True)
        else:  # the numerical values
            db[column].fillna(db[column].median(), inplace=True)
print("\nMissing values after filling:")
print(db.isnull().sum())
```

```python
# i will divide the age column into 5 bins(categories)
numerical_column = 'Age' # IT MUST BE NUMERICAL VALUE
bins = pd.cut(db[numerical_column], bins=5)
bin_counts = bins.value_counts().sort_index()
print(f"Binning of '{numerical_column}' into 5 equal-width bins:")
for interval, count in bin_counts.items():
    print(f"Range {interval}: {count} records")
```

```python
# i chose the maximum age in the age (Numerical column) and print its whole row
numerical_column = 'Age'
max_value_row = db.loc[db[numerical_column].idxmax()]
print(f"Row with the maximum value of '{numerical_column}':")
print(max_value_row)
```

```python
# Performing a boxplot on the weight column
boxplot_column = 'Weight'
plt.figure(figsize=(8, 6))
plt.boxplot(db[boxplot_column], vert=True, patch_artist=True)
plt.title(f"Boxplot of {boxplot_column}")
plt.ylabel(boxplot_column)
plt.grid(True)
plt.show()
# i chose it based on weight as it helps in visualizing the distribution of numbers and  to identify the outliers in the results
```
